# Distribution Plots


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_csv("D:\\Placement\\Data science notebook\\05-Seaborn\\dm_office_sales.csv")

plt.figure(figsize=(8,4), dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)
plt.title('Rug Plot')
plt.savefig('myRugplot.jpg')
```

```python

plt.figure(figsize=(12,4),dpi=200)
sns.displot(data=df,x='salary',bins=20,color='yellow', edgecolor='red', lw=2,ls='--')
sns.set(style='darkgrid')
plt.savefig('myhistoplot1.jpg')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.displot(data=df,x='salary',color='orange',kde=True,rug=True,alpha=0.6);
plt.savefig('myhiskderugplot.jpg')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.kdeplot(data=df, x='salary',fill=True)
plt.savefig('mykdeplot.jpg')
plt.title('KDE plot',loc='center')
```

# Distribution within Categories
1. Boxplot
2. Violinplot
3. Swarmplot
4. Boxenplot

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\adity\\Data science notebook\\05-Seaborn\\StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.boxplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set1')
plt.legend(bbox_to_anchor=(1.05,1))
plt.savefig('myBoxplot.jpg')
```

```python
plt.figure(figsize=(12,10),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set1')
plt.legend(bbox_to_anchor=(1.15,1))
plt.savefig('myViolinplot.jpg')
```

```python
plt.figure(figsize=(12,10),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2',split=True,inner='quartile')
plt.legend(bbox_to_anchor=(1.15,1))
plt.savefig('myViolinplotqwithsplit.jpg')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',hue='test preparation course',size=4)
plt.legend(bbox_to_anchor=(1.2,1))
plt.savefig('mySwarmplot1.jpg')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',hue='test preparation course',size=4,dodge=True)
plt.legend(bbox_to_anchor=(1,1))
plt.savefig('mySwarmplot2.jpg')
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.boxenplot(data=df,x='math score',y='test preparation course',hue='gender',palette='Set1')
plt.legend(bbox_to_anchor=(1.25,1))
plt.savefig('myBoxenplot.jpg')
```

# Comparison Plots-
1. Jointplot
2. Pairplot
3. Catplot
4. Pairgrid

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\adity\\Data science notebook\\05-Seaborn\\StudentsPerformance.csv")
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.jointplot(data=df, x='math score',y= 'reading score',hue='gender',palette='Set1')
plt.savefig('myjointplot.jpg')
```

```python
sns.pairplot(data=df,hue='gender')
plt.savefig('mypairplot1.jpg')
```

```python
sns.pairplot(data=df,hue='gender',corner=True)
plt.savefig('mypairplot2.jpg')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='strip',col='lunch', row='race/ethnicity')
plt.savefig('mycatplot.jpg')
```

```python
g=sns.PairGrid(df,hue='gender')
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.kdeplot)
g=g.map_lower(sns.kdeplot)
g=g.add_legend()
plt.savefig('myPairgrid.jpg')
```
